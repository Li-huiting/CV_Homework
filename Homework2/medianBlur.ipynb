```python
import cv2
import numpy as np
import math

def medianBlur(img, kernel, padding_way):
    #  img & kernel is List of List; padding_way a string
    kernel = np.array(kernel)
    k_h,k_w = kernel.shape
    img = np.array(img)
    img_h = img.shape[0]
    img_w = img.shape[1]
    output=np.zeros_like(img)
    
    # 检查padding_way的输入
    padding_way = padding_way.upper()
    padding_ways = {'ZERO', 'REPLICA'}
    if padding_way not in padding_ways:
        raise ValueError("Error Padding WAY! 'replica' & 'zero' ONLY!!")

   #填充图像
    pad_h = int((k_h - 1) / 2)
    pad_w = int((k_w - 1) / 2)
    if padding_way == 'REPLICA':
        if len(img.shape) == 3:
            img_pad = np.pad(img, ((pad_h, pad_h),(pad_w, pad_w), (0, 0)), 'edge')
        else:
            img_pad = np.pad(img, ((pad_h, pad_h),(pad_w, pad_w)), 'edge')
    if padding_way == 'ZERO':
        if len(img.shape) == 3:
            img_pad = np.pad(img, ((pad_h, pad_h),(pad_w, pad_w), (0, 0)), mode='constant', constant_values=0)
        else:
            img_pad = np.pad(img, ((pad_h, pad_h),(pad_w, pad_w)), mode='constant', constant_values=0)

    #取中值生成新图象
    for h in range(img_h):
        for w in range(img_w):
            window = img_pad[h:h+k_h-1,w:w+k_w-1] 
            output[h][w] = np.median(window,axis=(0,1))
    
    return output

img = cv2.imread('/home/lihuiting/Homework2/lena.jpg')
#print(img)
cv2.imshow('lena', img)
key = cv2.waitKey()
if key == 27:
    cv2.destroyAllWindows()

kernel = np.zeros((3,3))
img_median = medianBlur(img, kernel, 'ZERO')
#print(img_median) 
cv2.imshow('lena', img_median)
key = cv2.waitKey()
if key == 27:
    cv2.destroyAllWindows()

img_median = medianBlur(img, kernel, 'REPLICA')
#print(img_median) 
cv2.imshow('lena', img_median)
key = cv2.waitKey()
if key == 27:
    cv2.destroyAllWindows()
```
